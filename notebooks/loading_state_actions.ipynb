```python
import pandas as pd
import numpy as np
```

```python
state_actions = pd.read_csv('../data/state_action.csv').sample(frac=1)
```

```python
state_actions.head(12)
```

```python
actions = state_actions.iloc[:, -1:]
states = state_actions.iloc[:, :-1]
```

```python
states.shape
```

```python
board_temp = states.iloc[0].as_matrix().reshape((11,9,16))
```

```python
def print_layer(board, k):
    layer = np.dsplit(board, board.shape[2])[k].reshape((board.shape[0], board.shape[1]))
    print(layer)
```

```python
for i in range(16):
    print_layer(board_temp, i)
```
